```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Configuraci√≥ est√®tica
sns.set_theme(style="whitegrid")

# ==============================================================================
# DETECCI√ì AUTOM√ÄTICA DE LA RUTA (Per si tens el notebook a una subcarpeta)
# ==============================================================================

# Definim les rutes m√©s habituals on podria estar l'Excel
possibles_rutes = ['../data/raw/dades_prova.xlsx']

ruta_trobada = None

print(f"üìÇ El teu notebook est√† treballant des de: {os.getcwd()}")
print("üîç Buscant l'Excel...")

for ruta in possibles_rutes:
    if os.path.exists(ruta):
        ruta_trobada = ruta
        break

# ==============================================================================
# C√ÄRREGA I GR√ÄFICA
# ==============================================================================

if ruta_trobada:
    print(f"‚úÖ EUREKA! Fitxer trobat a: {ruta_trobada}")
    
    try:
        # Llegim l'Excel
        df = pd.read_excel(ruta_trobada)
        
        print(f"üìä Dades carregades: {df.shape[0]} files.")
        display(df.head()) # Mostra la taula
        
        # Fem la gr√†fica de prova
        plt.figure(figsize=(10, 6))
        sns.scatterplot(
            data=df,
            x='Ingresos',
            y='Consum',
            hue='Vulnerable',
            palette={0: 'green', 1: 'red'},
            s=100
        )
        plt.title("Prova Local: Ingressos vs Consum")
        plt.show()
        
    except Exception as e:
        print(f"‚ùå Error llegint l'Excel (potser et falta una llibreria?): {e}")

else:
    print("\n‚ùå NO TROBO EL FITXER.")
    print("Assegura't que l'estructura de carpetes √©s correcta.")
    print(f"He buscat a: {possibles_rutes}")
```
