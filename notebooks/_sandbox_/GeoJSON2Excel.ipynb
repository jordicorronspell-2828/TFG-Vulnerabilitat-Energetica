```python
import json
import pandas as pd
import os

# 1. Defineix la ruta del fitxer que t'has descarregat
# (Canvia 'nom_del_teu_arxiu_descarregat.geojson' pel nom real de l'arxiu)
# Fixa't en la "r" minúscula davant de les cometes
ruta_geojson = r'C:\Users\jordi\Documents\TFG_Projecte\TFG-Vulnerabilitat-Energetica\data\raw\CVI__2023-01-01.geojson'
print("Llegint l'arxiu GeoJSON...")

# 2. Obrim i carreguem el fitxer sencer a la memòria
with open(ruta_geojson, 'r', encoding='utf-8') as f:
    dades_geo = json.load(f)

# 3. Extraiem NOMÉS la part de les dades ("properties") de cada secció censal
llista_propietats = []
for feature in dades_geo['features']:
    llista_propietats.append(feature['properties'])

# 4. Ho convertim a un DataFrame de Pandas
df_crb = pd.DataFrame(llista_propietats)

# 5. Exportació a la ruta especificada
ruta_sortida_excel = '../data/processed/01_crb_dades_extretes.xlsx'
ruta_sortida_csv = '../data/processed/01_crb_dades_extretes.csv'

# Creem la carpeta si no existeix per evitar errors
os.makedirs('../data/processed/', exist_ok=True)

df_crb.to_excel(ruta_sortida_excel, index=False)
df_crb.to_csv(ruta_sortida_csv, index=False) # Et guardo també el CSV per si de cas

print(f"✅ Conversió completada! S'han extret dades de {len(df_crb)} seccions.")
print(f"📁 L'Excel s'ha guardat correctament a: {ruta_sortida_excel}")
print(f"\nLes columnes disponibles són:\n{df_crb.columns.tolist()}")
```
